## Import the data and build the training Data set
Some of the training tracks have lot of missing Data, so we choose to delete the trackes with the  whole line which have the 'None' or 'NaN' item. This step narrows the size of data and make the data matrix full filled.

```python
import pandas as pd
import numpy as np

# import the data
with open ("./data_in_matrixForm/trainIdx2_matrix.txt",'rt') as scoreFile:
    ScoreList = pd.read_csv(scoreFile,sep='|') # Every item(track album genre) with score

with open ("./YahooMusic/trackData2.txt",'rt') as trackFile:
    TrackList = pd.read_csv(trackFile,sep='|') # map to show every track belong which album and Genre.
    ### Delete the lines which have None ###
    album_noNone = TrackList['albumID'] != "None"
    artist_noNone = TrackList['artistID'] != "None" 
    album_noNaN = TrackList['albumID'] != np.NaN
    artist_noNaN = TrackList['artistID'] != np.NaN 
    genre_noNaN = TrackList['genreID'] != np.NaN
    TrackList = TrackList[album_noNone]
    TrackList = TrackList[artist_noNone]
    TrackList = TrackList[album_noNaN]
    TrackList = TrackList[artist_noNaN]
    TrackList = TrackList[genre_noNaN]
    
```

```python
# slice of the scorelist to find every user's track score.
trainingSet = ScoreList.loc[ScoreList['item'].isin(TrackList['trackID'])]
SortedList = trainingSet.sort_values(by=['user','score']) # sort the slice
```

```python
### Build the training data Set
def scoreFind(user,item):
    '''è¾“å…¥ç”¨æˆ·idï¼Œä»¥åŠæƒ³è¦æ‰¾çš„itemåï¼Œåœ¨ScoreListé‡Œå¯»æ‰¾å¯¹åº”çš„åˆ†æ•°/ input user ID and item ID, return the item's score '''
    n = ScoreList[ScoreList['user'] == user ]
    # if cannot find this itemï¼ˆAlbumï¼Œectï¼‰ï¼Œreturn -1.
    try:
        w = n[n['item'] == item]
        w['user'].iloc[0]
    except IndexError:
        return False
    else:
        score = w.iloc[:,2].iloc[0]
        return score

# Build a pandas dataframe
trainingScoreData = pd.DataFrame(columns=['user', 'track', 'album', 'artist', 'genre'])
# we only catch one genre for each track
k = 0
### The data we catch need have scores for all these items -- userï¼Œtrackï¼Œalbumï¼Œartistï¼Œgenre ###
### If any is None then we drop it ###
# Here we only choose 1000 user to make the training data set
for i in range(1,1000):
#     print(i)
    try:
        singleUser = SortedList[SortedList['user']==i] # single user's track score slice.
        singleUser['user'].iloc[0] # test if this user's list is empty raise IndexError.
    except IndexError:
        #print("user %d is not in this track based list" % i) # User is not in the Track only ScoreList
        continue
    for trackNumber in singleUser['item']:# track number
#         print(trackNumber)
        try:
            albumNumber=int(TrackList[TrackList['trackID']==trackNumber].iloc[:,1].iloc[0])
        except (IndexError, ValueError):
            #print("the album %d is not in the TrackList" % trackNumber)
            continue
            
        try:
            artistNumber=int(TrackList[TrackList['trackID']==trackNumber].iloc[:,2].iloc[0])
        except (IndexError, ValueError):
            #print("the artist %d is not in the TrackList" % trackNumber)
            continue
        
        try:
            genreNumber=int(TrackList[TrackList['trackID']==trackNumber].iloc[:,3].iloc[0])
        except (IndexError,ValueError):
            #print("the genre %d is not in the TrackList" % trackNumber)
            continue
        
        album = scoreFind(i,albumNumber)
        if album == False:
            continue
        artist = scoreFind(i,artistNumber)
        if not artist:
            continue
        genre = scoreFind(i,genreNumber)
        if not genre:
            continue
        trackScore = singleUser[singleUser['item']==trackNumber].iloc[:,2].iloc[0]
        trainingScoreData.loc[k]=[i,trackScore,album,artist,genre]
        #print(k)
        k += 1
```

```python
# Output Training Data
trainingScoreData.to_csv('./trainingData.csv',sep='|')
```

```python
# matrix_factorization
import numpy as np

def matrixFactorization(R,K,steps,alpha,beta):
    (row,col) = R.shape
    e = np.zeros((row,col))
    R_near = np.zeros((row,col))
    nP = P = np.random.random((row,K))
    nQ = Q = np.random.random((K,col))
    for n in range(1,steps+1):
        for i in range(row):
            for j in range(col):
                if R[i][j] != -1.0:
                    R_near = P.dot(Q)
                    e[i][j]=R[i][j]-R_near[i][j]
                    for m in range(K):
                        nP[i][m]=P[i][m]+alpha*(2*e[i][j]*Q[m][j]-beta*P[i][m])
                        nQ[m][j]=Q[m][j]+alpha*(2*e[i][j]*P[i][m]-beta*Q[m][j])
                        P[i][m] = nP[i][m]
                        Q[m][j] = nQ[m][j]
    return nP.dot(nQ)


with open('./YahooMusic/scores.txt','rt') as scorefile:
    scoreMatrix = np.loadtxt(scorefile,dtype=np.float128,delimiter="|")
    
filledMatrix=np.ones(shape=(94290,4))
i = 0
while i <90000 :
    R = scoreMatrix[i:i+5000,2:6]
    K = 2
    steps = 5
    alpha = 0.000001
    beta = 0.000002
    filledMatrix[i:i+5000,0:4]=matrixFactorization(R, K, steps, alpha, beta)
    i += 5000
    print (i)
    
R = scoreMatrix[90000:94290,2:6]
filledMatrix[90000:94290,0:4]= matrixFactorization(R, K, steps, alpha, beta)

# only repalce the '-1' item, keep the original items which have scores.
for i in range(0,94290):
    for j in range(0,4):
        if scoreMatrix[i][j+2] != -1.0:
            filledMatrix[i][j] = scoreMatrix[i][j+2]
            
```
